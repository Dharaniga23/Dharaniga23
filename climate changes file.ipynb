```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

```

```python
df = pd.read_csv('/content/climate_change_dataset.csv')
```

```python
print("Shape of dataset:", df.shape)
print("\nDataset columns:", df.columns)
print("\nFirst 5 rows:\n", df.head())
```

```python
print("\nMissing values:\n", df.isnull().sum())
```

```python
plt.figure(figsize=(12,6))
sns.lineplot(x='Year', y='Population', data=df, color='orange')
plt.title('Global Land Average Temperature Over Time')
plt.xlabel('Year')
plt.ylabel('Temperature (째C)')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
if 'LandAndOceanAverageTemperature' in df.columns:
    plt.figure(figsize=(12,6))
    sns.lineplot(x='Rainfall (mm)', y='Forest Area (%)', data=df, color='blue')
    plt.title('Global Land and Ocean Average Temperature Over Time')
    plt.xlabel('Year')
    plt.ylabel('Temperature (째C)')
    plt.xticks(rotation=45)
    plt.grid(True)
    plt.show()
```

```python

df_clean = df.dropna(subset=['Population'])

print("Shape after cleaning:", df_clean.shape)
```

```python

yearly_temp = df_clean.groupby('Year')['Renewable Energy (%)'].mean().reset_index()

print(yearly_temp.head())
```
